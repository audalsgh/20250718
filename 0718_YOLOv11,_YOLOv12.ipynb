미션1. 첫번째 미션 코드에서 YOLOv11로 바꿔보고,<br>
결과영상 다운로드 오류를 해결해보고자 했음.

```python
!pip install ultralytics

from google.colab import files
from ultralytics import YOLO # COCO 데이터셋에서 사전 훈련된 YOLOv11n 모델 로드

# 모델 로드후, 정보 표시 (선택사항)
model = YOLO("yolo11n.pt")
model.info()

# 동영상 업로드 후 추론 시작
print("동영상을 선택하세요")
uploaded = files.upload()
video_path = list(uploaded.keys())[0]
results = model(video_path, save=True)

# 영상 다운을 위한 라이브러리들
import os
import glob

# 📂 결과 영상 파일 경로 탐색
result_folders = glob.glob("runs/detect/predict2*")
if result_folders:
    latest_folder = max(result_folders, key=os.path.getctime)

    # ✅ 자동으로 해당 폴더에서 가장 최근에 저장된 avi 파일을 찾아서 다운로드
    result_mp4s = glob.glob(os.path.join(latest_folder, '*.avi'))
    if result_mp4s:
        result_mp4s.sort(key=os.path.getmtime, reverse=True)  # 최신 파일 우선
        result_video_path = result_mp4s[0]

        print(f"✅ 실제 저장된 결과 영상: {result_video_path}")
        files.download(result_video_path)
    else:
        print("❌ 결과 폴더에 mp4 파일이 없습니다.")
else:
    print("❌ 결과 영상이 저장되지 않았습니다.")
```

미션3. 첫번째 미션 코드에서 YOLOv8, YOLOv11, YOLOv12를 선택하여 사용하도록 바꿔보고,<br>
세가지 모두 사용해보았으니, 각각의 fps 차이를 기록해두어 다른 모델과 비교하기.<br>
+결과영상 다운로드 오류를 해결함.

```python
# 1. 설치
!pip install -q ultralytics opencv-python

# 2. 라이브러리 임포트
import cv2, time, os
from ultralytics import YOLO
from google.colab import files

# 3. 모델 선택
print("YOLO 모델 선택: yolo8n.pt / yolo11n.pt / yolo12n.pt")
model_name = input("모델 이름 입력 (예: yolo11n.pt): ").strip()

# 4. 모델 로드
try:
    model = YOLO(model_name)
    print(f"✅ {model_name} 모델 로드 성공")
except Exception as e:
    print(f"❌ 모델 로드 실패: {e}")
    raise SystemExit

# 5. 동영상 업로드
print("🎥 추론할 동영상을 업로드하세요")
uploaded = files.upload()
video_path = list(uploaded.keys())[0]
print(f"📂 업로드 완료: {video_path}")

# 6. 영상 열기 및 저장 설정
cap = cv2.VideoCapture(video_path)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps_input = cap.get(cv2.CAP_PROP_FPS)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_path = f"result_{model_name.split('.')[0]}.mp4"
out = cv2.VideoWriter(output_path, fourcc, fps_input, (w, h))

# 7. 추론 및 FPS 측정
print(f"🚀 {model_name} 추론 시작...")
start_time = time.time()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    annotated = results[0].plot()
    out.write(annotated)

cap.release()
out.release()
end_time = time.time()

# 8. FPS 계산 및 출력
elapsed_time = end_time - start_time
fps_measured = frame_count / elapsed_time

print(f"✅ 추론 완료: {output_path}")
print(f"📊 모델: {model_name} | 프레임 수: {frame_count} | 소요 시간: {elapsed_time:.2f}초 | 평균 FPS: {fps_measured:.2f}")

# 9. 결과 영상 다운로드
files.download(output_path)
```
