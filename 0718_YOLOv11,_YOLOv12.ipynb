ÎØ∏ÏÖò3. Ï≤´Î≤àÏß∏ ÎØ∏ÏÖò ÏΩîÎìúÏóêÏÑú YOLOv8, YOLOv11, YOLOv12Î•º ÏÑ†ÌÉùÌïòÏó¨ ÏÇ¨Ïö©ÌïòÎèÑÎ°ù Î∞îÍøîÎ≥¥Í≥†,<br>
ÏÑ∏Í∞ÄÏßÄ Î™®Îëê ÏÇ¨Ïö©Ìï¥Î≥¥ÏïòÏúºÎãà, Í∞ÅÍ∞ÅÏùò fps Ï∞®Ïù¥Î•º Í∏∞Î°ùÌï¥ÎëêÏñ¥ Îã§Î•∏ Î™®Îç∏Í≥º ÎπÑÍµêÌïòÍ∏∞.<br>
+Í≤∞Í≥ºÏòÅÏÉÅ Îã§Ïö¥Î°úÎìú Ïò§Î•òÎ•º Ìï¥Í≤∞Ìï®.

```python
# 1. ÏÑ§Ïπò
!pip install -q ultralytics opencv-python

# 2. ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏûÑÌè¨Ìä∏
import cv2, time, os
from ultralytics import YOLO
from google.colab import files

# 3. Î™®Îç∏ ÏÑ†ÌÉù
print("YOLO Î™®Îç∏ ÏÑ†ÌÉù: yolo8n.pt / yolo11n.pt / yolo12n.pt")
model_name = input("Î™®Îç∏ Ïù¥Î¶Ñ ÏûÖÎ†• (Ïòà: yolo11n.pt): ").strip()

# 4. Î™®Îç∏ Î°úÎìú
try:
    model = YOLO(model_name)
    print(f"‚úÖ {model_name} Î™®Îç∏ Î°úÎìú ÏÑ±Í≥µ")
except Exception as e:
    print(f"‚ùå Î™®Îç∏ Î°úÎìú Ïã§Ìå®: {e}")
    raise SystemExit

# 5. ÎèôÏòÅÏÉÅ ÏóÖÎ°úÎìú
print("üé• Ï∂îÎ°†Ìï† ÎèôÏòÅÏÉÅÏùÑ ÏóÖÎ°úÎìúÌïòÏÑ∏Ïöî")
uploaded = files.upload()
video_path = list(uploaded.keys())[0]
print(f"üìÇ ÏóÖÎ°úÎìú ÏôÑÎ£å: {video_path}")

# 6. ÏòÅÏÉÅ Ïó¥Í∏∞ Î∞è Ï†ÄÏû• ÏÑ§Ï†ï
cap = cv2.VideoCapture(video_path)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps_input = cap.get(cv2.CAP_PROP_FPS)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_path = f"result_{model_name.split('.')[0]}.mp4"
out = cv2.VideoWriter(output_path, fourcc, fps_input, (w, h))

# 7. Ï∂îÎ°† Î∞è FPS Ï∏°Ï†ï
print(f"üöÄ {model_name} Ï∂îÎ°† ÏãúÏûë...")
start_time = time.time()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    annotated = results[0].plot()
    out.write(annotated)

cap.release()
out.release()
end_time = time.time()

# 8. FPS Í≥ÑÏÇ∞ Î∞è Ï∂úÎ†•
elapsed_time = end_time - start_time
fps_measured = frame_count / elapsed_time

print(f"‚úÖ Ï∂îÎ°† ÏôÑÎ£å: {output_path}")
print(f"üìä Î™®Îç∏: {model_name} | ÌîÑÎ†àÏûÑ Ïàò: {frame_count} | ÏÜåÏöî ÏãúÍ∞Ñ: {elapsed_time:.2f}Ï¥à | ÌèâÍ∑† FPS: {fps_measured:.2f}")

# 9. Í≤∞Í≥º ÏòÅÏÉÅ Îã§Ïö¥Î°úÎìú
files.download(output_path)
```
